## How does the ReLU dataset behave (i.e. look) as a function of the hidden layer size k?

# k = 10

```python
fig, ax = plt.subplots(4, 4, figsize=(14, 14))
for i in range(4):
    for j in range(4):
        dim = 4
        n = 500
        k = 10
        points, labels = layer_relu_data(dim, n, k)
        ones = labels.astype(np.bool)
        zeros = np.invert(ones)
        ax[i, j].scatter(points[:, 0][ones], points[:, 1][ones], label='1')
        ax[i, j].scatter(points[:, 0][zeros], points[:, 1][zeros], label='0')
plt.show()
```
